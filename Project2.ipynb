```python
df['DateofHire']= pd.to_datetime(df['DateofHire'])
df['DateofHire'].head()
```

```python
today = pd.to_datetime('2023-07-01')
df['YearsofExperience'] = (today - df['DateofHire']) / np.timedelta64(1, 'Y')
df['YearsofExperience'].head()
```

```python
fig = plt.figure(figsize=(8,5))
g = sns.lmplot(x = 'EngagementSurvey', y = 'Salary', data = df)
g.fig.suptitle('Relationship between years of experience and salary', y = 1.05)
```

```python
df_group = df.groupby('Department').sum()
df_group = df_group.reset_index()
print(df_group)
```

```python
fig = plt.figure(figsize=(8,5))
bar = sns.barplot(x = 'Department', y = 'SpecialProjectsCount', order = ['IT/IS', 'Software Engineering', 'Admin Offices', 'Production', 'Sales', 'Executive Office'], data = df_group)
bar.set_title('Special projects contribution from each department')
```

```python
fig = plt.figure(figsize=(8,5))
box = sns.boxplot(data = df, x = 'Department', y = 'Salary')
box.set_title('Department wise salary distribution')
```

```python
fig = px.pie(df, names='PerformanceScore')
fig.show()
```

```python
df_new = df[['Department', 'PerformanceScore']]
df_new['value'] = 1
df_new = df_new.groupby(['Department', 'PerformanceScore']).count()
df_new = df_new.reset_index()
print(df_new)
```

```python
Ndep = len(df_new.Department.unique())
fig, axs = plt.subplots(1, Ndep, figsize=(4*Ndep,4), subplot_kw={'aspect':'equal'})
for ax,(groupname,subdf) in zip(axs,df_new.groupby('Department')):
    ax.pie(subdf.value, labels=subdf.PerformanceScore)
    ax.set_title(groupname)
```
